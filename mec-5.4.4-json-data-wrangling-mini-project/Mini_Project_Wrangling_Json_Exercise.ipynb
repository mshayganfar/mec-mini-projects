## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# To have a better understanding of data

sample_json_df.info()
```

## 1) Top 10 countries with most projects

```python
top_ten_countires_with_most_projects = sample_json_df.groupby('countryname').count().sort_values(by="project_name", ascending=False)[:10]
top_ten_countires_with_most_projects['project_count'] = top_ten_countires_with_most_projects.project_name
top_ten_countires_with_most_projects['project_count']
```

## 2) Top 10 major project themes

```python
# First we create a DataFrame from the given json data using json_normalize() function:

project_theme_data_list = sample_json_df.mjtheme_namecode.to_list()

normalized_project_theme_list = []
for project_theme in project_theme_data_list:
    normalized_project_theme_list.append(json_normalize(project_theme))

project_theme_df = pd.concat(normalized_project_theme_list, ignore_index=True)
project_theme_df
```

```python
#Note: There were 122 missing names in project_theme_df which we remove them from our couting int the following code:

# Using groupby():
# project_theme_df[project_theme_df.name != ''].groupby('name').count().sort_values(by="code", ascending=False)[:10]

# Using value_counts():
project_theme_df[project_theme_df.name != ''].name.value_counts().sort_values(ascending=False)[:10]
```

## 3) Dataframe with filled missing names:

```python
# First, we create a DataFrame without the missing project names.

non_missing_names_df = project_theme_df[project_theme_df.name != '']

non_missing_names_df = non_missing_names_df.reset_index(drop=True)
print(non_missing_names_df)
```

```python
# Then, we merge two DataFrames to fill the missing project names.

filled_theme_df = pd.merge(project_theme_df, non_missing_names_df, on='code', suffixes=('_x','')).drop('name_x' , axis=1)

print(f'Number of missing project names: {filled_theme_df[filled_theme_df.name == ""].name.count()}')
```
